```python
# Analyze sales trends over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Amount', data=df)
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.show()
```

```python
# Analyze product performance
# Example: Sales by category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Amount', data=df)
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer demographics
print(df['ship-country'].value_counts())

# Purchase patterns
# For example, top-selling products
top_products = df['SKU'].value_counts().head(10)
print(top_products)

# Visualize purchase patterns
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar')
plt.title('Top Selling Products')
plt.xlabel('Product SKU')
plt.ylabel('Number of Sales')
plt.show()
```

```python
# Extract month and year from Date column
df['YearMonth']  = pd.to_datetime(df['Date'], errors='coerce')


# Aggregate sales by month
monthly_sales = df.groupby('YearMonth')['Amount'].sum()

# Visualize sales trends by month
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Year-Month')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df1 = df.select_dtypes(include=[int, float])
```

```python
# Calculate correlation matrix
corr_matrix = df1.corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
