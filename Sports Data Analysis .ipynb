```python
# Step 3: Visualization for Player Performance
# Comparing player performance attributes for a sample of players
# Selecting a sample of players
sample_players = data1.sample(5)
# Attributes to compare
performance_attributes = ['Overall', 'Crossing', 'Finishing', 'ShortPassing']
# Plotting player performance
plt.figure(figsize=(12, 8))
for i, player_row in sample_players.iterrows():
    plt.plot(performance_attributes, player_row[performance_attributes], marker='o', label=player_row['Name'])
plt.title('Comparison of Player Performance Attributes')
plt.xlabel('Performance Attribute')
plt.ylabel('Attribute Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_sorted_by_age = data1.sort_values(by='Age')
# Plotting players' performance over time based on age
plt.figure(figsize=(12, 8))
plt.plot(df_sorted_by_age['Age'], df_sorted_by_age['Overall'], marker='o', linestyle='-')
plt.title('Players Performance Over Time based on Age')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.grid(True)
plt.show()
```

```python
# Grouping data by Age and Nationality and calculating average Overall rating
avg_overall_by_age_nationality = data1.groupby(['Age', 'Nationality'])['Overall'].mean().reset_index()
# Selecting some specific nationalities for comparison
selected_nationalities = ['England', 'Spain', 'Germany', 'France', 'Italy']
# Filtering data for selected nationalities
filtered_data = avg_overall_by_age_nationality[avg_overall_by_age_nationality['Nationality'].isin(selected_nationalities)]
# Creating subplots for each nationality
fig, axs = plt.subplots(len(selected_nationalities), 1, figsize=(10, 8), sharex=True)
for i, nationality in enumerate(selected_nationalities):
    # Filtering data for the current nationality
    nationality_data = filtered_data[filtered_data['Nationality'] == nationality]
    axs[i].bar(nationality_data['Age'], nationality_data['Overall'], color='skyblue')
    axs[i].set_title(f'Average Overall Rating by Age - {nationality}')
    axs[i].set_xlabel('Age')
    axs[i].set_ylabel('Average Overall Rating')
plt.tight_layout()
plt.show()
```
