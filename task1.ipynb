```python
# 3. Bar Plots for AQI categories
plt.figure(figsize=(12, 8))
for i, col in enumerate(['aqi_category', 'co_aqi_category', 'ozone_aqi_category', 'no2_aqi_category', 'pm2.5_aqi_category']):
    plt.subplot(2, 3, i+1)
    sns.countplot(x=col, data=data1)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# 4. Histograms for AQI values
plt.figure(figsize=(12, 8))
for i, col in enumerate(['aqi_value', 'co_aqi_value\t', 'ozone_aqi_value', 'no2_aqi_value', 'pm2.5_aqi_value']):
    plt.subplot(2, 3, i+1)
    sns.histplot(data1[col], bins=20, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 5. Scatter Plots
plt.figure(figsize=(12, 8))
sns.pairplot(data1[['aqi_value', 'co_aqi_value\t', 'ozone_aqi_value', 'no2_aqi_value', 'pm2.5_aqi_value']])
plt.title('Scatter Plot for AQI values')
plt.show()
```

```python
correlation=integers.corr()
```

```python
# Plotting correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='country_name', y='pm2.5_aqi_value', data=data1)
plt.title('Distribution of PM2.5 AQI Values Across Countries')
plt.xlabel('Country')
plt.ylabel('PM2.5 AQI Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting the line graph
plt.figure(figsize=(10, 6))
sns.lineplot(data=data1, x='country_name', y='aqi_value',  marker='o', linestyle='-')
plt.title('AQI Value by Country and City')
plt.xlabel('City')
plt.ylabel('AQI Value')
plt.legend(title='Country')
plt.grid(True)
plt.tight_layout()
plt.show()
```
